```python
# Instalação de uma ferramenta para redirecionar a porta 80 para 11434
!apt-get install -y socat

import subprocess
import threading

def start_proxy():
    print("Iniciando redirecionamento da porta 80 para 11434...")
    # Isso permite que acessos ao IP na porta 80 caiam no Ollama
    !sudo socat TCP-LISTEN:80,fork,reuseaddr TCP:127.0.0.1:11434

# Rodar em background
threading.Thread(target=start_proxy, daemon=True).start()
print("Proxy configurado. Tente acessar via seu domínio agora.")
```

```python
import requests
import json

# Exemplo de como consultar a API localmente ou via túnel
def query_ollama(prompt, model="qwen2.5-coder:7b", url="http://127.0.0.1:11434"):
    endpoint = f"{url}/api/generate"
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False
    }

    try:
        response = requests.post(endpoint, json=payload)
        return response.json().get('response', 'Erro na resposta')
    except Exception as e:
        return f"Erro ao conectar: {str(e)}"

# Teste local
print("Testando resposta do modelo...")
print(query_ollama("Olá, quem é você?"))
```

```python
import requests
import socket

# Descobrindo o IP público para configuração no DNS
print("--- CONFIGURAÇÃO DE DNS ---")
print("No seu provedor, aponte o Registro A de 'ollama.directlinkgo.online' para o IP abaixo:")
!curl -s ipv4.icanhazip.com
print("--------------------------\n")

# URL configurada com seu domínio
EXTERNAL_URL = "http://ollama.directlinkgo.online"

def query_external_ollama(prompt, model="qwen2.5-coder:7b"):
    endpoint = f"{EXTERNAL_URL}/api/generate"
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False
    }

    print(f"Consultando modelo {model} via {endpoint}...")
    try:
        # Nota: Certifique-se que o localtunnel ou proxy está aceitando conexões deste host
        response = requests.post(endpoint, json=payload, timeout=60)
        if response.status_code == 200:
            return response.json().get('response')
        else:
            return f"Erro na API: Status {response.status_code} - {response.text}"
    except Exception as e:
        return f"Erro de conexão: {str(e)}"

# Teste da API com o novo domínio
resultado = query_external_ollama("Teste de conexão com o novo domínio.")
print("\nResposta do Modelo:")
print(resultado)
```
